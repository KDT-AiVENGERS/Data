### 현재가격에 대한 전처리(\\, ₩ 처리), 한,영 강의만 추출 및 csv로 저장

```python
col = ['대분류', '소분류', '강의명', '난이도', '가격(현재가격)', '가격(원래가격)',
       '총소요시간', '강의소개', '언어', '출처']
file_list = ['udemy_data-science.csv', 'udemy_databases.csv', 'udemy_development-tools.csv', 'udemy_game-development.csv',
             'udemy_mobile-apps.csv', 'udemy_no-code-development.csv', 'udemy_software-engineering.csv',
             'udemy_software-testing.csv', 'udemy_web-development.csv']
```

```python
for file in file_list:
    d = pd.read_csv(f'./udemy/{file}')
    d = d.query('언어 in ["한국어", "영어"]')[col]
    d['가격(현재가격)'] = d['가격(현재가격)'].str.replace('\\', '', regex = True)
    d['가격(현재가격)'] = d['가격(현재가격)'].str.replace('₩', '', regex = True)
    d.to_csv(f'./udemy/KE_{file}', index = False)
```

### 현재가격 숫자로 형식 통일 및 integer형으로 변환 (무료의 경우 0원으로 처리)  
### null이 존재하는 소분류, 강의소개 삭제  
### 총소요시간 'x시간 y분' 형태로 통일(온디맨드 = 주문형)  

```python
import pandas as pd
```

```python
pd.read_csv('./data/pre_result.csv')
```

```python
df = pd.read_csv('./udemy/udemy_resultKE.csv')
```

```python
def prep(df):
    df = df[~df['소분류'].isnull()]
    df = df[~df['강의소개'].isnull()]
    df= df.drop('가격(원래가격)', axis = 1)
    df['가격(현재가격)'] = df['가격(현재가격)'].str.replace(',', '')
    df['가격(현재가격)'] = df['가격(현재가격)'].str.replace('무료', '0')
    df['가격(현재가격)'] = df['가격(현재가격)'].astype('int')
    df['총소요시간'] = df['총소요시간'].str.replace(' 주문형 동영상', '')
    df['총소요시간'] = df['총소요시간'].str.replace('온디맨드 동영상 ', '')
    df.reset_index(drop = True, inplace = True)
    
    #소요시간 처리 ex. 10시간 30분
    for idx in range(len(df)):
        x = df['총소요시간'][idx]
        try:
            if '.5' in x:
                df.loc[idx, '총소요시간'] = x.replace('.5', '')+ ' ' + '30분'
        except:
            pass # nan data의 경우 float형이므로 iterable하지 않아 in 안에 들어가지 못하므로 예외처리 합니다.
    
    return df
```

```python
df = prep(df)
```
