```python
from collections import Counter
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv("./Data_result/pre_result_morph.csv")
```

```python
def s(x):
    try:
        return ast.literal_eval(x)
    except (ValueError, SyntaxError):
        return x

a = df.applymap(s)
print(a['주요업무'][5])
```

```python
test = a['주요업무']

re_list = []
for i,data in test.items():
    re_list.append(data)
```

```python
print(re_list[:5])
```

```python
new_list = sum(re_list,[])
```

```python
new_list_c = Counter(new_list)
print(new_list_c)
```
