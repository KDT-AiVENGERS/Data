```python
# 8,095개 데이터 크롤링 완료 (걸린 시간 : 7시간 1분 12초)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rc

df = pd.read_csv('크롤링_원티드_ver02_all_230627.csv')
df.head()
```

```python
# 컬럼별 null값 갯수 파악 : 기술스택/툴이 없는 회사가 많네요.
df.isnull().sum()
```

```python
# 5개의 데이터는 페이지에 접근 후, 오류가 발생했습니다.
# 아까 들어갔을 땐 '유효하지 않은 채용공고입니다'? 같은 오류 메세지만 떴었는데, 다시 들어가보니 제대로 뜨네요!
df[df['회사명'].isnull()]
```

```python
#데이터셋의 총 갯수는 8,095개입니다.
len(df)
```

```python
# 출처 기준으로 중복된 값을 세어봤는데 3203개..! 너무 많네요.
# 중복된 값 중에서 어떤 걸 남겨야할지는 회의를 통해 결정하는 게 좋을 것 같아 아직 처리하지 않았습니다.
df.duplicated(subset=['출처']).sum()
```

```python
# 직무별로 수집된 데이터 확인
df['직무명'].value_counts()
```

```python
# 직무별로 수집된 데이터 수를 내림차순하여 시각화해두었으므로 참고 부탁드립니다!
# 데이터 사이언티스트, DBA 등은 채용공고가 많은데도 데이터가 부족한 걸 보실 수 있는데요.
# 제 와이파이 or 서버 문제로 중간에 스크롤이 제대로 안 먹은 것 같아서, 추후 '출처' 중복체크하면서 재수집하면 좋을듯합니다.
rc('font', family='AppleGothic') 			
plt.rcParams['axes.unicode_minus'] = False 
plt.figure(figsize=(15,10))
job_count = df['직무명'].value_counts().sort_values(ascending=False)
sns.barplot(x=job_count.values, y=job_count.index)
# 막대 오른쪽 끝에 값을 표시하는 로직 추가
for i, v in enumerate(job_count.values):
    plt.text(v + 3, i, str(v), color='black', va='center')
```
