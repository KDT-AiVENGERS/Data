```python
'''
한국어, 영어인 강의들(유료+무료) 합쳐서 result라는 이름의 한 파일로 저장합니다.
categories = ["web-development", "data-science", "mobile-apps", "programming-languages", "game-development",
              "databases", "software-testing", "software-engineering", "development-tools", "no-code-development"]

'''
```

```python
import pandas as pd

df0 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_web-development.csv')
df2 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_data-science.csv')
df1 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_mobile-apps.csv')
df3 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_programming-languages.csv')
df4 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_game-development.csv')
df5 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_databases.csv')
df6 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_software-testing.csv')
df7 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_software-engineering.csv')
df8 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_development-tools.csv')
df9 = pd.read_csv('../../재현/udemy_csv/한국어,영어강의csv/KE_udemy_no-code-development.csv')
df_free = pd.read_csv('./free_lecture/KE_udemy_free.csv')

```

```python
df_0to9 = pd.concat([df0, df1, df2, df3, df4, df5, df6, df7, df8, df9], ignore_index=True)
df_0to9.isnull().sum()
```

```python
df_paid = df_0to9[df_0to9["가격(현재가격)"]!="무료"]
df_paid.isnull().sum()
```

```python
pd.set_option('display.max_colwidth', None)
# df_paid[df_paid['소분류'].isnull()] #소분류가 NaN인 값은 진짜 소분류가 없고 대분류까지만 있는 강의들
# 총소요시간, 강의소개는 진짜 없는 것들
# 이대로 저장합니다!
df_all = pd.concat([df_paid, df_free], ignore_index=True)
df_all
```

```python
df_all.to_csv("./udemy_resultKE.csv", index=False)
```
