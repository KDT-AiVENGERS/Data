```python
'''
23.07.27(목)
Capitalize(첫글자만 대문자) 적용하면 단어 단위로 분리가 되어야 하므로, "소문자"로 통일하기로 했음.
사용자 사전에 "소문자"로 변환한 기술 스택의 단어들을 넣어줄 예정.
원본 pre_result 데이터의 모든 영어를 소문자로 바꾸는 작업은 customOkt돌릴 때 추가하기로 함.
'''
import pandas as pd
df = pd.read_csv('/Users/lora/Documents/GitHub/Data/찬혁/Data_Pretreatment/Data_result/pre_wanted.csv')['기술 스택']
```

```python
# lower로 변경하는 함수 적용
def convert_to_lower(text):
    if isinstance(text, str):
        return text.lower()
    else:
        return text
df_lower = df.apply(convert_to_lower)
```

```python
skill_list = df_lower.str.split('\n').explode().tolist()
skill_list
```

```python
# 중복제거 후 결과 출력
unique_values = list(set(skill_list))
print(unique_values)
```

```python
#test
a = str(unique_values[2]).split(' ')
a
```

```python
# 띄어쓰기 없애주기
unique_nospace = []
for value in unique_values:
    unique_nospace.append(str(value).split(' '))
unique_nospace = sum(unique_nospace,[])
unique_nospace = list(set(unique_nospace)) # 중복제거
unique_nospace
```

```python
import csv
# CSV 파일 이름
csv_file = "skill_nospace.csv"

# 데이터를 CSV 파일로 저장
with open(csv_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["단어명"])
    for value in unique_nospace:
        writer.writerow([value])
```
